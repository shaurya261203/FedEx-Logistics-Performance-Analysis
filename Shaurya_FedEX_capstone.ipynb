#### Chart - 3

````python
#data cleaned to get accurate analysis

df['delivery delay'] = (df['Delivered to Client Date']-df['Scheduled Delivery Date']).dt.days # extracted day difference between scheduled and actual delivery date
df.loc[df['delivery delay'] > 0, 'delivery delay'].sum() # selected days which took more or less than 0 days to deliver from the scheduled date
Q1 = df['delivery delay'].quantile(0.25)
Q3 = df['delivery delay'].quantile(0.75)

#performed IQR calculation to detect the accurate outliers
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df[(df['delivery delay'] >= lower) & (df['delivery delay'] <= upper)] #selected the data between lower bound and upper bound , rest everything not considered.

print(df_clean['delivery delay'].value_counts())#printed the count of days on which delivery has been reported.
print(df_clean.loc[df['delivery delay']<0,'delivery delay'].value_counts().sum())#printed the total number of delayed deliveries which is 621.
````

````python
# Chart - 3 visualization code : (Box Plot)
sns.boxplot(df_clean,x='Shipment Mode',y='delivery delay',color='black')

plt.title('Delivery Delay with Shipment Mode')
plt.xlabel('Shipment Mode')
plt.ylabel('Delivery Delay')
plt.tight_layout()
plt.show()
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code(heatmap)
top_countries = df.groupby('Country')['Line Item Value'].sum().nlargest(15).index#grouping by with country and value and adding it and giving 15 largest value for data
df_top = df[df['Country'].isin(top_countries)]


heatmap_data = df_top.pivot_table(index='Country', columns='Product Group', values='Line Item Value',  aggfunc='sum') # creating a pivot table for creating a corelation in a chart
heatmap_data = heatmap_data / 1000000 # Scaling to Millions

plt.figure(figsize=(14, 10))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="YlOrRd", cbar_kws={'label': 'Total Value ($M)'})#a heatmap format cmap=defines the color ,annot= puts data inside the box

plt.title('Regional Demand Heatmap: Country vs. Product Group ($ Millions)', fontsize=16)
plt.xlabel('Product Group', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.show()
````

##### 1. Why did you pick the specific chart?

I selected the heatmap because it clearly shows the relationship between country and product group in one view.

##### 2. What is/are the insight(s) found from the chart?

from the heatmap we can see which countries contribute the highest spending and which product groups are most demanded in each region. some countries show strong concentration in specific product groups while others have more balanced demand. this helps in understanding regional preference and identifying key markets driving overall revenue.
